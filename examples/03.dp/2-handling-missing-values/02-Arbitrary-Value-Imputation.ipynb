## Arbitrary value imputation - pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# We'll use the following variables,
# 3 of which contain NA.
cols_to_use = [
    "OverallQual",
    "TotalBsmtSF",
    "1stFlrSF",
    "GrLivArea",
    "WoodDeckSF",
    "BsmtUnfSF",
    "LotFrontage",
    "MasVnrArea",
    "GarageYrBlt",
    "SalePrice",
]
```

```python
# Let's load the House Prices dataset.
fileUrl = 'https://raw.githubusercontent.com/cmps460s26/cmps460-content/refs/heads/main/examples/03.dp/datasets/houseprice.csv'
df = pd.read_csv(fileUrl, usecols=cols_to_use)

print(df.shape)
df.head()
```

Split the data set into a training and a testing set.

```python
# Let's separate into training and testing sets.

X_train, X_test, y_train, y_test = train_test_split(
    df.drop("SalePrice", axis=1),
    df["SalePrice"],
    test_size=0.3,
    random_state=0,
)

X_train.shape, X_test.shape
```

```python
# Find missing data
X_train.isnull().mean()
```

```python
# Capture the variables to impute in a list.
vars_to_impute = [var for var in X_train.columns if X_train[var].isnull().sum() > 0]

vars_to_impute
```

```python
# Let's plot the distributions of the variables.

X_train[vars_to_impute].hist(bins=50, figsize=(10, 8))
plt.show()
```

```python
X_train[vars_to_impute].agg(["min", "max"])
```

```python
# Capture the imputation values in
# a dictionary

imputation_dict = {
    "LotFrontage": 999,
    "MasVnrArea": 1999,
    "GarageYrBlt": 2999,
}

imputation_dict
```

```python
# Replace missing data

X_train.fillna(imputation_dict, inplace=True)
X_test.fillna(imputation_dict, inplace=True)
```

```python
# Corroborate replacement

X_train.isnull().sum()
```

```python
# Ensure imputation was done

X_test.isnull().sum()
```
